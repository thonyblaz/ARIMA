```python
import numpy as np
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt

# Datos originales
data = [30, 21, 29, 25, 30, 22, 26, 24, 28, 23, 27, 25]
# Convertir a una serie temporal
data_series = pd.Series(data)

# Ajustar el modelo ARIMA(3,1,3)
model = ARIMA(data_series, order=(3, 1, 3))
model_fit = model.fit()

# Resumen del modelo
print(model_fit.summary())
```

```python
# Predicci√≥n de 3 valores futuros
forecast = model_fit.forecast(steps=10)
print(forecast)
```

```python
# Graficar datos originales y predicciones
plt.figure(figsize=(10, 5))
plt.plot(data_series, label='Datos Originales')
plt.plot(range(len(data_series), len(data_series) + 3), forecast, label='Predicciones', color='red')
plt.legend()
plt.show()
```
